### Problema 6 Modelo de regresiÃ³n Lineal

```python
def regresion_lineal(x,y,epochs, imprimir_error_cada, alpha):

    vect = np.ones_like(x).reshape(-1,1)
    x = np.reshape(x,(-1,1))
    A = np.hstack((x,vect))

    b1 = 0
    b0 = 0

    beta = np.array([b1,b0], dtype=np.float64)
    

    for epoch in range(epochs):

        y_pred = np.dot(A, beta)

        error = np.mean((y-y_pred)**2)

        if epoch % imprimir_error_cada == 0:
            print(f"IteraciÃ³n {epoch}, Error = {error:.4f} ")

        delta_bo = np.mean(y_pred - y)
        delta_b1 = np.mean(np.dot(y_pred,x))

        beta -= alpha * np.array([delta_b1,delta_bo]) 

    return beta     
        

```

```python
beta = regresion_lineal(entrenamiento["overallqual"].to_numpy(), entrenamiento["Sales_Price"].to_numpy(), 10000, 100, 0.0001)
print("beta0 = {:.2f}, beta1 = {:.2f}".format(beta[1], beta[0]))
```
