### Problema 6 Modelo de regresión Lineal

```python
def regresion_lineal(x, y, epochs, imprimir_error_cada, alpha):
    vect = np.ones_like(x).reshape(-1, 1)
    x = np.reshape(x, (-1, 1))
    A = np.hstack((x, vect))

    beta = np.array([0.0, 0.0])
    modelo = {}
    errores = []

    for epoch in range(epochs):
        
        y_pred = np.dot(A, beta)
        error = (np.mean((y - y_pred)**2) )/ 2
        errores.append(error)

        if epoch % imprimir_error_cada == 0:
            print(f"Iteración {epoch}, Error = {error:.4f}")

        grad = np.dot(A.T, (y_pred - y)) / len(x)
        beta -= alpha * grad
        modelo[epoch] = beta.copy()

    return errores, modelo   
        

```

```python
errores, modelo = regresion_lineal(entrenamiento["overallqual"].values, entrenamiento["Sales_Price"].values, 100, 10, 0.001)
for i, b in modelo.items():
    print(f"Iteración {i}, beta0 = {b[0]:.2f}, beta1 = {b[1]:.2f}")


```

### Gráfica de errores 

```python
def graficar_errores(iteraciones, errores):
    plt.scatter(iteraciones, errores)
    plt.title("Error en las iteraciones")
    plt.xlabel("Iteraciones")
    plt.ylabel("Errores")
    plt.show()

iteraciones_errores = regresion_lineal(entrenamiento["overallqual"].values, entrenamiento["Sales_Price"].values, 100, 10, 0.001)
errores = iteraciones_errores[0]
iteraciones = np.arange(len(errores))
graficar_errores(iteraciones, errores)
```

### Gráfica de puntos

```python
def visualizar_modelo(modelos, x, y, n):

    fig, ax = plt.subplots()
    ax.scatter(x, y)
    ax.set_xlabel('OverallQual')
    ax.set_ylabel('Sales_Price')
    
    for i in range(0, len(modelos), n):
        beta = modelos[i]
        x_rango = np.linspace(x.min(), x.max())
        y_rango = beta[0] * x_rango + beta[1]
        ax.plot(x_rango, y_rango, label=f'Iteración {i}')
    
    ax.legend()
    plt.title("Evolución de los betas de la regresión Lineal")
    plt.show()


betas_guardadas = regresion_lineal(entrenamiento["overallqual"].values, entrenamiento["Sales_Price"].values, 100, 10, 0.001)
visualizar_modelo(modelos= betas_guardadas[1], x=entrenamiento['overallqual'], y=entrenamiento['Sales_Price'], n=10)
```
